# Introduction
The iris data is famous dataset used for machine learning and data analysis.The objective of this task is to explore the dataset, understand its structure, and visuaslize the data usinf different graphs such as scatter plot,histogram and box plot.

```python
import pandas as pd
import seaborn as sns
import matplotlib .pyplot as plt
```

```python
df=sns.load_dataset("iris")
df.head()
```

```python
print(df.shape)
print(df.columns)
df.head()
```

The dataset contains 150 rows and 5 columns.
the columns include sepal length,sepal width,petal length,petal width and species.

```python
df.isnull(). sum()
df.duplicated().sum()
```

```python
df.info()
df.describe()
```

```python
sns.scatterplot(x="sepal_length",y="petal_length",hue="species",data=df)
plt.title("sepal length vs petal length")
plt.show()
```

The scatter plot shows the relationship between sepal length and petal length.Setosa species is clearly separated from the other species.

```python
plt.hist(df["sepal_length"])
plt.title("Distribution of sepal length")
plt.xlabel("sepal length")
plt.ylabel("frequency")
plt.show()
```

The histogram shows the distribution of sepal length values.Most values lies between 5 and 7.

```python
sns.boxplot(x="species",y="sepal_length",data=df)
plt.title("Boxplot of sepal length by species")
plt.show()
```

The boxplot shows the spread of sepal length for each species and helps identify outliers.
